### Kaggle API token

Run the cell below, then click **Choose Files** and select `kaggle.json` (download from <https://www.kaggle.com/settings> → 'Create New API Token').

If you've already configured `~/.kaggle/kaggle.json` in this Colab session, you can skip this cell.

```python
import os, pathlib, shutil
kaggle_dir = pathlib.Path.home() / '.kaggle'
kaggle_json = kaggle_dir / 'kaggle.json'
if not kaggle_json.exists():
    from google.colab import files
    print('Upload your kaggle.json:')
    uploaded = files.upload()
    if 'kaggle.json' not in uploaded:
        raise SystemExit('No kaggle.json was uploaded.')
    kaggle_dir.mkdir(parents=True, exist_ok=True)
    with open(kaggle_json, 'wb') as fp:
        fp.write(uploaded['kaggle.json'])
    os.chmod(kaggle_json, 0o600)
    print(f'Saved to {kaggle_json}')
else:
    print(f'Already configured: {kaggle_json}')
```
